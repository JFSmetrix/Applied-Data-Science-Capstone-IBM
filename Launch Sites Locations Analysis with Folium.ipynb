```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

# Function to calculate Haversine distance between two sets of coordinates
def haversine_distance(lat1, lon1, lat2, lon2):
    # Radius of the Earth in kilometers
    R = 6371.0
    
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Calculate differences in coordinates
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Distance in kilometers
    distance = R * c
    return distance

# Create a distance matrix for launch sites
launch_sites_matrix = {}

for index, site1 in launch_sites_df.iterrows():
    distances = []
    for index, site2 in launch_sites_df.iterrows():
        if site1['Launch Site'] != site2['Launch Site']:
            distance = haversine_distance(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
            distances.append(distance)
        else:
            distances.append(0.0)  # Distance to itself is 0
    launch_sites_matrix[site1['Launch Site']] = distances

# Display the distance matrix
launch_sites_distances_df = pd.DataFrame(launch_sites_matrix, index=launch_sites_df['Launch Site'])
launch_sites_distances_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
